```python
def gender(x):
    if x == 'male':
        return 1
    else:
        return 0
```

```python
df['new_gen'] = df['Sex'].apply(gender)
```

```python
df.head()
```

```python
def fare_cat(n):
    if n > 300:
        return 'First Class'
    elif n > 200:
        return 'Second Class'
    else:
        return 'Third Class'
```

```python
df['New_Fare_cat'] = df['Fare'].apply(fare_cat)
```
